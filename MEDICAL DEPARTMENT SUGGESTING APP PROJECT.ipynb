```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler(feature_range=(49,50))
rescaler=scaler.fit_transform(x)
```

# TRAIN TEST SPLIT

```python

from sklearn.model_selection import StratifiedKFold
fold=StratifiedKFold(n_splits=25,random_state=1,shuffle=True)
fold
```

```python
for train_index,test_index in fold.split(x,data.DEPARTMENT):
    xtrain,xtest,ytrain,ytest=x[train_index],x[test_index],data.DEPARTMENT[train_index],data.DEPARTMENT[test_index]
```

# TRAIN TEST FIT

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.naive_bayes import GaussianNB
```

```python
model_params={
   "svm":{
       "model":svm.SVC(gamma="auto"),
       "params":{"C":[5,39,25],
                "kernel":["rbf","linear"]
       
   } },
    "random_forest":{
        "model":RandomForestClassifier(),
        "params":{
            "n_estimators":[5,10,100]
    } },
    "logistic_regression":{
        "model":LogisticRegression(),
        "params":{
            "C":[5,10,24]
    } },
  }
```

```python
resultscores=[]
for model_name in model_params:
    mp=model_params[model_name]
    gds=GridSearchCV(mp["model"],mp["params"],cv=5,return_train_score=True)
    gds.fit(x,y)
    resultscores.append({"model":model_name,"best_score":gds.best_score_,
                  "best_params":gds.best_params_})
```

```python
df=pd.DataFrame(resultscores,columns=["model","best_score","best_params"])
df
```

```python
MEDICAL_APP=RandomForestClassifier(n_estimators=10)
MEDICAL_APP.fit(xtrain,ytrain)
```

```python
MEDICAL_APP.score(xtest,ytest)
```

# MAKING PREDICTIONS

```python
y_pred=MEDICAL_APP.predict(xtest)          #predictions
y_pred
```

# CHECKING ACCURACY

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest,y_pred)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(ytest,y_pred)
plt.figure(figsize=(5,3))
seaborn.heatmap(cm,annot=True)
```
